#### <i><b><span style='color:#55679C'>Example</span> : Show the customer_id, total_revenue and total_revenue_after_tax columns in the telco_services table, where the total_revenue_after_tax column is the total_revenue minus the 10% tax payable! #### <i><b><span style='color:#55679C'>Example</span></b></i>
The company wants to apply certain discounts with the following conditions :
<ul>
    <li>If <code>tenure_in_months</code> under one year, then give a 2% discount on the customer's monthly fee</li>
    <li>If <code>tenure_in_months</code> between 1 - 3 years, then give a 5% discount on the customer's monthly fee</li>
    <li>If <code>tenure_in_months</code> over 3 years, then give a 10% discount on the customer's monthly fee</li>
</ul><br>
Present customer_id, tenure_in_months, monthly_charge, discount (charge cut not percentage) and monthly_charge_after_disc!

```python
SELECT
  customer_id,
  tenure_in_months,
  monthly_charge,
  CASE
    WHEN tenure_in_months < 12 THEN ROUND((0.02) * monthly_charge, 2)
    WHEN tenure_in_months BETWEEN 12 AND 36 THEN ROUND((0.05) * monthly_charge, 2)
    ELSE ROUND((0.10) * monthly_charge, 2)
  END AS discount,
   CASE
    WHEN tenure_in_months < 12 THEN ROUND((1-0.02) * monthly_charge, 2)
    WHEN tenure_in_months BETWEEN 12 AND 36 THEN ROUND((1-0.05) * monthly_charge, 2)
    ELSE ROUND((1-0.10) * monthly_charge, 2)
  END AS monthly_charge_after_disc
FROM telco_churn.telco_services
```

#### <i><b><span style='color:#55679C'>Example</span> : is a metric used to measure the level of customer loyalty to a brand or company. This metric can help companies understand how strong their relationship is with their customers..</b></i><br>
Customer Loyalty Score can be formulated as :
\begin{equation}
\text{Customer Loyalty Score (CLS)} = \text{Customer Satisfaction Score} \times \frac{\text{CLTV}}{100}
\end{equation}

In `customer_status` table, show customer_id, cltv, satisfaction_score, cls and status!

```python
SELECT
  customer_id,
  cltv,
  satisfaction_score,
  satisfaction_score * (cltv/100) AS customer_loyalty_score,
  status
FROM telco_churn.customer_status
```

#### <i><b><span style='color:#55679C'>Quest</span> : </b></i><br>
Customer Loyalty Score can be formulated as :
\begin{equation}
\text{Customer Loyalty Score (CLS)} = \text{Customer Satisfaction Score} \times \frac{\text{CLTV}}{100}
\end{equation}

In `customer_status`, Present customer_id, cltv, satisfaction_score, cls and status columns. But round cls with rounding without any numbers behind the comma (ROUND(0)) and display only those with churn status then sort by largest to smallest cls!

```python
SELECT
  customer_id,
  cltv,
  satisfaction_score,
  ROUND(satisfaction_score * (cltv/100), 0) AS customer_loyalty_score,
  status
FROM telco_churn.customer_status
WHERE status = 'Churned'
ORDER BY customer_loyalty_score DESC
```

# <b>C. <span style='color:#0B2F9F'><code>Date Data Type Operation</code></span></b>

In SQL, The date data type is used to store or operate date data. The date data type can perform the following operations, including:
<ol>
    <li>Present current date and time</li>
    <li>Add or Subsract date</li>
    <li>Getting the date part</li>
    <li>Calculating date differences</li>
</ol>

<br><b>NOTE</b><br><br>
Things to note when using operations `date` is :
<ul>
    <li>The standard SQL date format is 'YYYY-MM-DD' or year - month - date, for example: 2024-10-14 shows the date October 14, 2024</li>
    <li>Accessing dates in SQL requires using single quotes, for example: '2024-10-14'</li>
    <li>Make sure the value or date column data type is DATE or TIMESTAMP, if not then it can be converted to <code>DATE(tanggal)</code> atau <code>TIMESTAMP(tanggal)</code></li>
</ul>

<i>docs : <a href="https://cloud.google.com/bigquery/docs/reference/standard-sql/date_functions">https://cloud.google.com/bigquery/docs/reference/standard-sql/date_functions</a></i>

#### <i><b><span style='color:#55679C'>Example</span> : Current date is January 1, 2025 and tenure_in_months is a column in the telco_services table, then for each customer_id determine the estimated date each customer starts subscribing</b></i><br>

```python
SELECT
  customer_id,
  DATE_SUB(DATE('2025-01-01'), INTERVAL tenure_in_months MONTH) AS join_date,
  tenure_in_months
FROM telco_churn.telco_services
```

#### <i><b><span style='color:#55679C'>Example</span> : For example, if the current date is January 1, 2025 and tenure_in_months is a column in the telco_services table, then for each customer_id, determine the estimated month (in the month name format) & year each customer started subscribing.!</b></i><br>

```python
SELECT
  customer_id,
  FORMAT_DATE('%B %Y', DATE_SUB(DATE('2025-01-01'), INTERVAL tenure_in_months MONTH)) AS join_month_year,
  tenure_in_months
FROM telco_churn.telco_services
```
